```python
import numpy as np
import cv2 as cv
from skimage import io
import matplotlib.pyplot as plt
from scipy import fftpack
```

```python
src = cv.imread('noisy2.tif',0)
plt.imshow(src, cmap='gray')
height = np.size(src, 0)
width = np.size(src, 1)
```

```python
final = cv.medianBlur(src, 15)
```

```python
plt.imshow(final, cmap='gray')
```

```python
kernel = np.ones((7,7),np.uint8)
img_erosion = cv.erode(final, kernel, iterations=1) 
```

```python
plt.imshow(img_erosion, cmap='gray')
```

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv.dilate(img_erosion,kernel,iterations = 2)
plt.imshow(dilation, cmap='gray')
```

```python
kernel = np.ones((23,23),np.uint8)
closing = cv.morphologyEx(dilation, cv.MORPH_CLOSE, kernel)

plt.imshow(closing, cmap='gray')
```

```python
dst =  src * closing
plt.imshow(dst, cmap='gray')
```
