```python
import numpy as np
import cv2 as cv
from skimage import io
import matplotlib.pyplot as plt
```

```python
src = cv.imread('sap.jpg',0)
```

```python
plt.imshow(src, cmap='gray')
```

```python
dst = cv.medianBlur(src, 5)
```

```python
plt.imshow(dst, cmap='gray')
```

```python
res = np.hstack((src,dst))
```

```python
plt.imshow(res, cmap='gray')
```
