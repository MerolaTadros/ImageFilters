```python
import numpy as np
import cv2 as cv
from skimage import io
import matplotlib.pyplot as plt
from scipy import fftpack
```

```python
src = cv.imread('noisy2.tif',0)
plt.imshow(src, cmap='gray')
height = np.size(src, 0)
width = np.size(src, 1)
```

```python
thresh0 = cv.adaptiveThreshold(src, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV, 15, 70)
plt.imshow(thresh0, cmap='gray')
```

```python
kernel = np.ones((2,1),np.uint8)
opening = cv.morphologyEx(thresh0, cv.MORPH_OPEN, kernel)
```

```python
plt.imshow(opening, cmap='gray')
```

```python
kernel = np.ones((3,3),np.uint8)
dilation = cv.dilate(opening,kernel,iterations = 2)
plt.imshow(dilation, cmap='gray')
```

```python
kernel = np.ones((25,25),np.uint8)
closing = cv.morphologyEx(dilation, cv.MORPH_CLOSE, kernel)
plt.imshow(closing, cmap='gray')
```
